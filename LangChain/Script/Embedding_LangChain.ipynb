```python
from langchain.embeddings import OpenAIEmbeddings
```

```python
openai_embed = OpenAIEmbeddings(openai_api_key = "your_api_key")
text = "Welcome to LangChain, happpy learning"

embedded = openai_embed.embed_query(text)

print(embedded)
```

Embedding the Words from HuggingFace

```python
from langchain.embeddings import HuggingFaceEmbeddings

text_new = "Welcome to LangChain, happpy learning"

huggingface_model_pulling = HuggingFaceEmbeddings(model_name = "sentence-transformers/all-MiniLM-L6-v2")
```

```python
embedded_new = huggingface_model_pulling.embed_query(text_new)

print(embedded_new)
```
