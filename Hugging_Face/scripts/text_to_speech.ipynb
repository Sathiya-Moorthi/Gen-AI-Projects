### Understanding `Audio()` vs `display(Audio())` in Jupyter Notebooks

There are two ways to show audio widgets in Jupyter notebooks:

1. Direct Usage: `Audio(data, rate=sample_rate)`
   - Works when it's the last expression in a cell
   - Jupyter automatically displays the last expression
   - This is why it worked in the earlier cell

2. Inside a Loop or Mid-cell: `display(Audio(data, rate=sample_rate))`
   - Required when Audio is not the last expression
   - Needed inside loops or conditional statements
   - Explicitly tells Jupyter to display the widget
   - This is why we need it in the loop

Example:
```python
# This works (last expression):
Audio(data)

# This doesn't work in a loop:
for i in range(2):
    Audio(data)  # Won't display

# This works in a loop:
for i in range(2):
    display(Audio(data))  # Will display
```

The `display()` function is a special IPython command that forces output to be shown, regardless of where it appears in the cell.

````python
from IPython.display import Audio

inputs = processor("Hello, my name is Suno. And, I am a text-to-speech model. Welcome Sathiya to Suno!")

speech_output = model.generate(**inputs)

Audio(speech_output[0].numpy(), rate=model.generation_config.sample_rate)
````
